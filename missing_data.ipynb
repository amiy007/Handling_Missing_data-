```python
import pandas as pd
import numpy as np;
dataset=pd.read_csv('diabetes.csv');
dataset
```

```python
print((dataset[["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]]==0).sum())

```

```python
dataset[["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]]=dataset[["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]].replace(0,np.NaN)
```

```python
dataset
```

```python
print(dataset.isnull().sum())
```

```python
# Removing rows with missing values using dropna
dataset.dropna(inplace=True)
print(dataset.shape)
#removed rows with missing values cutting from 768 to 392 rows and same no. of columns i.e 9
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
#split dataset into inputs and outputs
values=dataset.values
x=values[:,0:8]
y=values[:,8]

#evaluate an LDA model on the dataset usindg k-fold cross validation
model=LinearDiscriminantAnalysis()
kfold=KFold(n_splits=3, random_state=7)
result=cross_val_score(model,x,y,cv=kfold, scoring='accuracy')
print(result.mean())
  
# prints accuracy of model 
```
